## How to Find Specific Information in the Dataset with df.loc

```python
df.loc[df["Sex"] == "female"]
```

```python
df.loc[(df["Sex"] == "female") & (df["Pclass"] == 1)]
```

```python
df.loc[(df["Sex"] == "female") & (df["Pclass"] == 1) & (df["Survived"] == 0)]
```
